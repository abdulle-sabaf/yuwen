```python
path = '../src/小学/小学第一册.tex'

with open(path, 'r', encoding="utf-8") as f:
    lines = f.readlines()
```

```python
firsts_list = []
articles_list = []

buff = 1
last_title = ""

for line in lines:
    if line.startswith("\section") or line.startswith("\end"):
        title = line.split("{")[-1].split("}")[0]
        if isinstance(buff, str):
            if last_title in ('document', "第二十四课"):
                articles_list.append(buff)
            elif last_title:
                firsts_list.append(buff)
        buff = ""
        last_title = title
    elif isinstance(buff, str):
        buff += line
```

```python
def sort_article(art):
    lines = art.split("\n")
    title = ""
    content = []
    para = ""
    for line in lines:
        if line:
            if not title:
                title = line
            else:
                para += line
        elif para:
            content.append(para)
            para = ""

    return title, content

title, content = sort_article(articles_list[8])
```

```python
def keys(dct):
    return list(dct.keys())
```

```python
articles = {}

for art in articles_list:
    title, content = sort_article(art)
    if title == "古诗两首":
        continue
    articles[title] = content
```

```python
def get_prop(shi, inputs, output_keys=("idx",), verbose=0):
    # inputs: dict 
    # output_keys: iterable of key strings
    # verbose: 0: no output, 1: output the errors in outputting, 2: output everything
    for k, v in inputs.items():
        if k in shi and shi[k] == v:
            continue
        else:
            if verbose > 1:
                print(f"Mismatch in [{shi['idx']}] input. key: {k}, [{v}] <--> [{shi[k]}].")
            return {}
    if "all" in output_keys:
        return shi.copy()
    outputs = {}
    for k in output_keys:
        if k in shi:
            outputs[k] = shi[k]
        elif verbose:
            print(f"Error in [{shi['idx']}] output. key: {k} not found.")
    return outputs

def lookup_prop(shis, inputs, output_keys=("all",), verbose=0, mode="first"):
    results = []
    for shi in shis:
        r = get_prop(shi, inputs, output_keys=output_keys, verbose=verbose)
        if r:
            if mode == "first":
                return r
            elif mode == "all":
                results.append(r)
    if mode == "last":
        return r
    return results

for i, (beg, fin) in enumerate(marks):
    beg_found = False
    fin_found = False
    for shi in shis:
        r = get_prop(shi, beg)
        if r:
            print(i+1, "begin:", r["idx"])
            beg_found = True
        r = get_prop(shi, fin)
        if r:
            print(i+1, "end:", r["idx"])
            fin_found = True
    if not beg_found:
        print("begin:", beg, "not found")
    if not fin_found:
        print("end:", fin, "not found")
```

```python
marks_idx = []
for i, (beg, fin) in enumerate(marks):
    beg_idx, fin_idx = -1, -1
    r = lookup_prop(shis, beg)
    if r:
        beg_idx = int(r["idx"]) - 1
    r = lookup_prop(shis, fin)
    if r:
        fin_idx = int(r["idx"])
    marks_idx.append((beg_idx, fin_idx))

for i, (beg, fin) in enumerate(marks_idx):
    genre = genres[i]
    print(genre, beg, fin)
    for j in range(beg, fin):
        shis[j]["genre"] = genre
```
